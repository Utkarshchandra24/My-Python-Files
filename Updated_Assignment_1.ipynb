```python
#1.Slicing Strings
b = "Hello, World!"
print(b[2:5])

#1.1Slice From the Start
b = "Hello, World!"
print(b[:5])

#1.2 Slice To the End
b = "Hello, World!"
print(b[2:])

#1.3 Negative Indexing
b = "Hello, World!"
print(b[-5:-2])

#2.Modify Strings
#2.1 Upper Case
a = "Hello, World!"
print(a.upper())

#2.2 Lower Case
a = "Hello, World!"
print(a.lower())

#2.3 Remove Whitespace
a = " Hello, World! "
print(a.strip()) 

#2.4 Replace String
a = "Hello, World!"
print(a.replace("H", "J"))

#2.5 Split String
a = "Hello, World!"
print(a.split(",")) 

#3.Concatenation String
a = "Hello"
b = "World"
c = a + b
print(c)

#4.Format String
age = 36
txt = "My name is John, and I am {}"
print(txt.format(age))

quantity = 3
itemno = 567
price = 49.95
myorder = "I want {} pieces of item {} for {} dollars."
print(myorder.format(quantity, itemno, price))

```
